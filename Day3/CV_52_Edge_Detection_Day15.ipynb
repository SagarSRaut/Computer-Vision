## Sobel Kernel (Gradient based)

````python
# Load the image
image = cv2.imread(r"/content/drive/MyDrive/Colab Notebooks/images/lenna.png",0) # gray image

# Sobel parameters: ksize, dx,dy, ddepth
sobelx = cv2.Sobel(image,cv2.CV_64F,1,0,ksize=3) # sobel along horizontal edges
sobely = cv2.Sobel(image,cv2.CV_64F,0,1,ksize=3)

# combined sobel
combined_sobel = cv2.add(sobelx, sobely)

thresh, binary_sobel = cv2.threshold(combined_sobel, 100, 255, cv2.THRESH_BINARY)


# show the images
cv2_imshow(image)
cv2_imshow(sobelx)
cv2_imshow(sobely)
cv2_imshow(binary_sobel)
````

## Superimposing images

````python
img1 = cv2.imread(r"/content/drive/MyDrive/Colab Notebooks/images/lenna.png")
img2 = cv2.imread(r"/content/drive/MyDrive/Colab Notebooks/images/sun.png")
````

````python
img1.shape, img2.shape
````

````python
new_img = cv2.add(img1,img2)  #superimposing one image over the another, condition is that the size of the image arrays has to be same
final_image = np.hstack((img1,img2,new_img))
cv2_imshow(final_image)
````

## Canny Edge detection
