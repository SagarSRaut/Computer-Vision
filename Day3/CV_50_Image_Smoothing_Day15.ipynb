## Averaging kernel

```python
# Load the image
image = cv2.imread(r"/content/drive/MyDrive/Colab Notebooks/images/lenna.png")

# Define a 5x5 averaging kernel
kernel = np.ones((5,5), np.float32) / 25

# Apply the kernel to the image
smoothed_image = cv2.filter2D(image, -1, kernel)

# Display the original and smoothed images
# cv2_imshow(image)
# cv2_imshow(smoothed_image)

#horizontally stacking the original and new image
cv2_imshow(np.hstack((image, smoothed_image)))
```
