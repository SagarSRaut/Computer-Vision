## Averaging kernel

```python
# Load the image
image = cv2.imread(r"/content/drive/MyDrive/Colab Notebooks/images/lenna.png",0) # gray image

# define a threshold
binary_image = np.where(image < 127, 0, 255)

# show the image
# cv2_imshow(binary_image)
np.hstack((cv2_imshow(image),cv2_imshow(binary_image)))

## Making darker values dark and whiter values white
```

## Inverse Binary Thresholding

```python
# define a threshold
binary_image = np.where(image > 127, 0, 255) #replace all pixel values which are greater than 127 with 0 and the rest with 255

# show the image
# cv2_imshow(binary_image)
np.hstack((cv2_imshow(image),cv2_imshow(binary_image)))

## Making darker values white and whiter values dark
```

## Otsu threshold

```python
ret, thresh1 = cv2.threshold(image,120,255,cv2.THRESH_BINARY) # if value < 120 make it 0 else make it 255
ret, thresh2 = cv2.threshold(image,120,255,cv2.THRESH_BINARY_INV) # if value > 120 make it 255 else make it 0

#show the original, binray and inverese binary images
np.hstack((cv2_imshow(image),cv2_imshow(thresh1),cv2_imshow(thresh2)))
```

```python
otsu_value, thresh_otsu = cv2.threshold(image,0,255,cv2.THRESH_OTSU) # Otsu thresholding
```
