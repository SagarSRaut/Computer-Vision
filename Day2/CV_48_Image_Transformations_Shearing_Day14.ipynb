```python
# Load the image
image = cv2.imread(r"/content/drive/MyDrive/Colab Notebooks/images/lenna.png")

# Defining shearing factors(adjust as needed)
shear_x = -0.25
shear_y = 0.45

#Create a transformation matrix
M = np.float32([[1, shear_x, 0], [shear_y, 1, 0], [0,0,1]])

rows, cols, _ = image.shape
T = np.float32( [[1, 0, 0], [0, 1, 150]])

# We use warpAffine to transform the image using the matrix, T
img_translation = cv2.warpAffine(image, T, (int(cols*1.5), int(rows*1.5)))
sheared_img = cv2.warpPerspective(img_translation, M, (int(cols*1.5), int(rows*1.5)))

cv2_imshow(image)
cv2_imshow(sheared_img)
```
