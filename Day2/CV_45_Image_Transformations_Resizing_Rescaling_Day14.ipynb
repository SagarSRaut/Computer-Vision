## Resize using scikit-image

```python
from skimage.transform import resize

# load the image
image = plt.imread(r"/content/drive/MyDrive/Colab Notebooks/images/lenna.png")

# Resize with bilinear interpolation
resized_image = resize(image, (1000,1200), anti_aliasing=True)

# Resize with bicubic interpolation
resized_image = resize(image, (1300,1300), anti_aliasing=True, order=3) #order ==> averaging the pixel values of the front 3 and the back 3 pixels

# show the resized image
plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original Image")

plt.subplot(1,2,2)
plt.imshow(resized_image)
plt.title("Resized Image")

plt.show()
```
