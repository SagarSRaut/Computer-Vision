### Using default rotation

```python
  from google.colab.patches import cv2_imshow
```

```python
# Load the image
image = cv2.imread(r"/content/drive/MyDrive/Colab Notebooks/images/lenna.png")

# Rotate the image 45 degrees clockwise
rotated_image = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE) # ROTATE_90_CLOCKWISE

# show the rotated image
cv2_imshow(rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Rotation using matrix

```python
# load image
image = cv2.imread(r"/content/drive/MyDrive/Colab Notebooks/images/lenna.png")

#Define the center point and rotation angle
angle = 60

# Create a rotation matrix
rotation_matrix = cv2.getRotationMatrix2D((255,255), angle, 1)

# Rotate the image
rotated_image = cv2.warpAffine(image, rotation_matrix, (768,768))

# show the rotated image
cv2_imshow(rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
