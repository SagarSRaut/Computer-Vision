```python
# Load the image
image = cv2.imread(r"/content/drive/MyDrive/Colab Notebooks/images/lenna.png")

# resize image to a fixed size
flipped_horizontal = cv2.flip(image, 0) # 0 for vertical, 1 for horizontal, -1 for both (180 degree)

flipped_Vertical = cv2.flip(image, 1) # 0 for vertical, 1 for horizontal, -1 for both (180 degree)

flipped_180 = cv2.flip(image, -1) # 0 for vertical, 1 for horizontal, -1 for both (180 degree)


# show the rotated image
cv2_imshow(flipped_horizontal)
cv2_imshow(flipped_Vertical)
cv2_imshow(flipped_180)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
